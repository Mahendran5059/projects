### Transform and manipulating data(Feature Engineering)

We created new column or property <i>ride_length</i> to measure total time for each rider _id by difference of ride ride ending and starting time 

```python
annual2022_clean["ride_length"]=annual2022_clean["ended_at"]-annual2022_clean["started_at"]
```

```python
ts=pd.to_timedelta("0 days 00:00:00")
```

```python
annual2022_clean[annual2022_clean["ride_length"]<=ts].shape
```

<br>
negative value of ride length is impossible. so, we clean the that error data

```python
annual2022_clean=annual2022_clean[~(annual2022_clean["ride_length"]<=ts)]
```

# Share Phase
share our findings using Visualizations

```python
mem_month=(annual2022_clean[annual2022_clean["member_casual"]=="member"].groupby("month")["ride_length"].mean())
```

```python
def to_minutes(td):
    return td.total_seconds() / 60
```

```python
def to_days(td):
    return td.total_seconds() / (60*60*24)
```

```python
mem_month.apply(to_minutes)
```

```python
casual_month=(annual2022_clean[annual2022_clean["member_casual"]=="casual"].groupby("month")["ride_length"].mean())
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=mem_month,x=mem_month.index,y=mem_month.apply(to_minutes),color="brown",alpha=0.7)
sns.lineplot(data=casual_month,x=casual_month.index,y=casual_month.apply(to_minutes),color="orange")
sns.scatterplot(data=mem_month,x=mem_month.index,y=mem_month.apply(to_minutes),color="black")
sns.scatterplot(data=casual_month,x=casual_month.index,y=casual_month.apply(to_minutes),color="red")
plt.ylim(0,30)
plt.xlim(0,13)
plt.ylabel("Average_Ride_Length_Minutes",color="olive",size=15)
plt.xlabel("Month",color="olive",size=15)
plt.title("Average_Ride_Length of Casual and Memer over Months",size=20,color='g')
plt.legend(["member","","casual","","member data points","casual data points"],loc="upper right")
plt.show()
```

Above graph shows the decreasing trend of average ride length of casual riders and seasonal changes in ride length of member riders, particularly ride increased in 5th, 6th, 7th month of 2022
