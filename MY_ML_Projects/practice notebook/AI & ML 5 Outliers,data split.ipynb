# Standardization
Data standardization is the process of rescaling one or more attribute so that they have a value of 0 and a standard deviation of 1.
Standardization assumes that your data has a Gaussian(bell curve) distribution.This does no strictly have to be true , but the technigue is more effective

```python
import pandas as pd
import numpy as np
np.abs(-89.9)
```

```python
def detect_outliers(data):
    outlier=[]
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
```
