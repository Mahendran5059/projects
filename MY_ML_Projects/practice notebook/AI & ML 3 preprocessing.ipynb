```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],'A':['A0','A1','A2','A3']})
right=pd.DataFrame({'key':['K0','K1','K2','K4'],'B':['B0','B1','B2','B4']})
pd.merge(left,right,how="left",on='key')
#outer= print all in both table
#inner=print common observations in both table
#left=print all from left T and common from right T
#right=print all from right and common from left T
```
